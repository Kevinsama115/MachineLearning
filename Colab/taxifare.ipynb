```python

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression


np.random.seed(42)

distance = np.linspace(1, 20, 50)
base_fare = 2000
per_km_rate = 1200


fare = base_fare + distance * per_km_rate + np.random.normal(0, 300, size=distance.shape)

df = pd.DataFrame({
    "distance_km": distance,
    "fare_krw": fare.astype(int)
})

print(df.head())

X = df[['distance_km']]
y = df['fare_krw']

model = LinearRegression()
model.fit(X, y)

pred = model.predict([[3.5]])
print(f"\nEstimated fare for 3.5 km: {pred[0]:.0f} KRW")

plt.scatter(X, y)
plt.plot(X, model.predict(X))

plt.xlabel("Distance (km)")
plt.ylabel("Fare (KRW)")
plt.title("Taxi Fare Prediction")

plt.show()
```

```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt

# 1. Create dataset
usage_time = np.array([1, 2, 3, 4, 5, 6]).reshape(-1, 1)
battery_used = np.array([10, 18, 30, 38, 50, 65])

# 2. Train model
model = LinearRegression()
model.fit(usage_time, battery_used)

# 3. Predict for 4 hours
pred = model.predict([[4]])
print(f"Predicted battery usage for 4 hours: {pred[0]:.2f}%")

# 4. Visualization
plt.scatter(usage_time, battery_used)
plt.plot(usage_time, model.predict(usage_time))

plt.xlabel("Usage Time (hours)")
plt.ylabel("Battery Used (%)")
plt.title("Phone Usage vs Battery Consumption")

plt.show()
```
